```python
#gdp per capita growth comparisions
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('countries.csv')
data
```

```python
india = data[data.country == 'India']
china = data[data.country == 'China']
mexico = data[data.country == 'Mexico']
fig, ax = plt.subplots(figsize=(10,8))
plt.plot(india.year, india.gdpPerCapita)
plt.plot(china.year, china.gdpPerCapita)
plt.plot(mexico.year, mexico.gdpPerCapita)
plt.title('GDP per Capita growth ', fontsize=18)
plt.legend(['India','China', 'Mexico'], prop={'size': 18})
ax.set_xlabel('Year', labelpad=15,fontsize=18)
ax.set_ylabel('GDP per capita in $.', labelpad=15,fontsize=18)
plt.show()
```

```python
india_growth = india.gdpPerCapita/india.gdpPerCapita.iloc[0]
china_growth = china.gdpPerCapita/china.gdpPerCapita.iloc[0]
mexico_growth = mexico.gdpPerCapita/mexico.gdpPerCapita.iloc[0]
fig,ax = plt.subplots(figsize=(10,8))
plt.title('GDP growth rate comparision')
plt.plot(india.year, india_growth)
plt.plot(china.year, china_growth)
plt.plot(mexico.year,mexico_growth)
plt.legend(['India','China','Mexico'], prop={'size': 18})
plt.xlabel('Year', fontsize=20)
plt.ylabel('GDP growth rate', fontsize=20)
plt.show()
```

# Comparing population growth in India and China

```python
india_pop = india.population/india.population.iloc[0]*100
china_pop = china.population/china.population.iloc[0]*100
bangladesh = data[data.country =='Bangladesh']
bangla_pop =bangladesh.population/bangladesh.population.iloc[0]*100
plt.figure(figsize=(10,8))
plt.title('Population Growth rate comparision')
plt.plot(india.year, india_pop)
plt.plot(china.year, china_pop)
plt.plot(bangladesh.year, bangla_pop)
plt.legend(['India','China','Bangladesh'], prop={'size':18})
plt.xlabel('Year', fontsize=18)
plt.ylabel('Population grwoth rate',fontsize=18)
plt.show()
```
